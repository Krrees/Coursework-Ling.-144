A Note on `join`
---------------

In pigLatinParts, you will naturally find yourself wanting to connect some tokens into a string again. Remember that we got tokens as a **list**, a sequence of strings, like this:

    li = ["a", "b", "c"]

Suppose I want to print out "a b c"; well in that case you can use what we did on Wednesday:

    print li[0], li[1], li[2]
    # you might want to compare that with print li, just to see what happens

But if I want to get a new variable that is "a b c", I can't print. What do I do?

One thing you can do is do it manually

    stringFromLi = li[0] + " " + li[1] + " " + li[2]

But what if I want something more general? In that case, you can use the nifty function `join`, which is designed to link elements of a list together, where the link is given by a string **delimiter**. Remember that I said the stuff between tokens was called a delimiter; well, the same reasoning applies here, since we are detokenizing, or as it's often called **stringifying**.

Now this is where Python's object-first view makes things rather confusing. In Perl, for instance, if I wanted to join a list with a delimiter, I'd write something basically like:

    print join(delimiter, list)

If that worked in Python, we'd write

    print join(" ", li)         # (1)
    
But it doesn't. `join` isn't a basic function like `len`. Rather, it's a [method of ALL strings](http://docs.python.org/2/library/stdtypes.html#str.join). jQuery202018406376136534375_1381259255889?? Well, the idea is that any string could be a delimiter for a list, so what we wanted to do in (1) is actually done this way:

    print " ".join(li)
    
I really really really hate how that looks. One way to make it look better is to set the delimiter to a variable (and not a *literal* space):

    delim = " "
    print delim.join(li)
    
But I still really dislike this. I think somehow that one is joining the list, not joining from the delimiter. If you dislike this as much as I do, you can roll your own `join` function:

    def join(delim, li):
        return delim.join(li)
    
If you put that in a program you are running, then you can safely call `join` in what I think is a more natural way. Try it out below:

```python
def join(delim, li):
    return delim.join(li)

li = ["a", "b", "c"]
print join(" THIS IS A SPACE, REALLY ", li)
```
